```python
from GA.visual.genetic import unsupervisedGeneticDraw
#from GA.visual.mutator import NaiveMutatorDraw
from GA.visual.plot import Plot
from GA.visual.util import createEllipsis,createImage
#from GA.general.probability import StaticProbability
from GA.general.component import Population
```

```python
initialImages=[]
for i in range(5):
    initialImages.append(createImage(id=i,shapes=createEllipsis(num=1000,ellipseSize=0.05)))
```

```python
population=Population(size=len(initialImages),generation=0,individuals=initialImages)
```

```python
#mutator=NaiveMutatorDraw()
p=Plot()
g=unsupervisedGeneticDraw(mutator=None,                          
                          population=population,
                          plot=p)
g.run(iter=5)
```
